### 2.2.1 A) What are the three lowest ages and three highest ages of mothers?

```python
#Drop NA

df3 = df3[pd.notnull(df3['dtnascmae'])]
```

```python
# Convert to Integer

df3['dtnascmae'] = df3['dtnascmae'].astype('int64')


# Extract Year

df3['anomaenasc'] = df3['dtnascmae'].astype(str).str[-4:].astype('int64')

df3['anomaenasc'] = df3['anomaenasc'].replace([992,  998, 982,  990,  976, 985,  984,  988, 981, 986, 993,  980,  999, 90,  991,  996,  979,  994, 997,  995,  987, 199, 989], 
                                              [1992, 1998, 1982, 1990, 1976, 1985, 1984, 1998, 1981, 1986, 1993,  1980,  1999, 1990,  1991,  1996,  1979,  1994, 1997,  1995,  1987, 1999, 1989])


#Extract Month
df3['mesmaenasc'] = df3['dtnascmae'].astype(str).str[2:-4].astype(int)

df3['mesmaenasc'] = df3['mesmaenasc'].replace([0], [10])



#Extract Day
df3['diamaenasc'] = df3['dtnascmae'].astype(str).str[:-6].astype(int)



#Create Colunm date
cols=['diamaenasc','mesmaenasc','anomaenasc']
df3['date'] = df3[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis=1)



#Convert to date
df3['date']=pd.to_datetime(df3['date'])



#Create Column date reference
day_calc_age = '2021-07-12'

df3['day_calc_age'] = day_calc_age

df3['day_calc_age'] = pd.to_datetime(df3['day_calc_age'], format='%Y-%m-%d')



#Subtract date
df3['idademae'] = (df3['day_calc_age'] - df3['date'])

#Convert to year age
df3['idademae'] = abs(df3['idademae'] / np.timedelta64(1, 'Y'))
df3['idademae'] = df3['idademae'].astype('int64')
```

```python
# datafram with age

'''Nesse ponto o erro encontrado nas idades, que possivelmente foi coletado de forma errada,
   ficando dificil mensurar ou imputar algum dado, resolvi desconsiderar as idades mais elevadas nas respostas finais'''

'''At this point, the error found in the ages, which was possibly collected incorrectly,
   making it difficult to measure or impute any data, deciding to disregard the higher ages in the final answers'''

#older ages
older = df3[df3['idademae'] < 60 ].sort_values('idademae', ascending=False)

older['idademae'].head()
```

```python
'''Da mesma forma ocorre com as menores idades, resolvi desconsiderar as idades menos elevadas nas respostas finais,
   considerando apenas as idades superiores a 13 anos'''

'''Likewise with younger ages, I decided to disregard younger ages in the final answers,
   considering only ages over 13'''

#younger ages


younger = df3[df3['idademae'] > 5 ].sort_values('idademae', ascending=True)

younger['idademae'].head()
```
